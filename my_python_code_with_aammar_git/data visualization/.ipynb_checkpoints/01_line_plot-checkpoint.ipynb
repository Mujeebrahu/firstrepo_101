```python
# import librareis
import seaborn as sns
import matplotlib.pyplot as plt

# lood dataset
phool=sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plat") #adding titles
plt.show()
```

## Adding limtis


```python
# import librareis
import seaborn as sns
import matplotlib.pyplot as plt

# lood dataset
phool=sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plat") #adding titles
plt.xlim(2) # add limits in you plot on x_axes
plt.ylim(1)# add limitts in your plot on y_axes
plt.show()
```

```python
# import librareis
import seaborn as sns
import matplotlib.pyplot as plt

# lood dataset
phool=sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plat") #adding titles
# plt.xlim(2) # add limits in you plot on x_axes
# plt.ylim(1)# add limitts in your plot on y_axes
# set style
sns.set_style("dark") ## in this line code "plot ka background bark kar rahe hen"
sns.set_style(style=None, rc=None) # in this line code "plot ka background bark kar rahe hen"
plt.show()
```

## size of figure/plot

```python
#size of figure/plot
# import librareis
import seaborn as sns
import matplotlib.pyplot as plt

# lood dataset
phool=sns.load_dataset("iris")
phool
#change figure/ploot
plt.figure(figsize=(4,2))
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plat") #adding titles

#

sns.set_style("dark")
plt.show()
```
